```sql
-- 1. Check existing catalogs
SHOW CATALOGS;

-- 2. Create the main project catalog
CREATE CATALOG IF NOT EXISTS german_energy_transition_lakehouse MANAGED LOCATION 'abfss://german1energy@stgerenergyarthidev.dfs.core.windows.net/'
COMMENT 'Catalog for German Energy Transition Analytics Lakehouse project';
```

```sql
-- 3. Select the catalog
USE CATALOG german_energy_transition_lakehouse;

-- 4. Create project schemas
CREATE SCHEMA IF NOT EXISTS landing;

CREATE SCHEMA IF NOT EXISTS bronze;

CREATE SCHEMA IF NOT EXISTS silver;

CREATE SCHEMA IF NOT EXISTS gold;

CREATE SCHEMA IF NOT EXISTS audit
COMMENT 'Audit schema for control tables, pipeline logs, and data quality results';

CREATE SCHEMA IF NOT EXISTS ml
COMMENT 'ML schema for feature tables and simple baseline model outputs';

-- 5. Check schemas
SHOW SCHEMAS IN german_energy_transition_lakehouse;
```

6. Create external volume for raw landing files


```sql
CREATE EXTERNAL VOLUME IF NOT EXISTS german_energy_transition_lakehouse.landing.raw_files
LOCATION 'abfss://german1energy@stgerenergyarthidev.dfs.core.windows.net/landing/'
COMMENT 'External volume for raw OPSD source files in the landing layer';
```

7. Check volume

```sql
SHOW VOLUMES IN german_energy_transition_lakehouse.landing;
```

```sql
LIST '/Volumes/german_energy_transition_lakehouse/landing/raw_files/';
```
